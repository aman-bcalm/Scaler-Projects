```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
from sklearn import preprocessing


import warnings
warnings.filterwarnings('ignore')
```

```python
dt = pd.read_csv('delhivery_data.csv')
```

```python
dt.columns[dt.isna().any()].tolist()
```

```python
#Conversion of categorical attributes to category

#Categorical variables are data, route_type

dt['data'] = dt['data'].map({'training':1, 'test':0})
dt['route_type'] = dt['route_type'].map({'FTL':1, 'Carting':0})
```

```python
#source_center, source_name, destination_center, destination_name have a lot of categories hence we cannot use
#one hot encoding as the number of columns will explode and it would become very difficult to analyze


dt["source_center"] = dt["source_center"].astype('category')
dict_sc =  dict( enumerate(dt['source_center'].cat.categories  ) )
dt["source_center"] = dt["source_center"].cat.codes
```

```python
#length of the dictionary is equal to the number of unique values in source_center hence the dict mapping is
#correct
len(dict_sc)
```

```python
#Now converting for other remaining variables 
```

```python
#desitnation_center
dt["destination_center"] = dt["destination_center"].astype('category')
dict_dc =  dict( enumerate(dt['destination_center'].cat.categories  ) )
dt["destination_center"] = dt["destination_center"].cat.codes
```

```python
#Did not convert Source name and Destination name into numerical values as we need to extract features out of it
```

```python
#converting from object to datetime

dt["trip_creation_time"]= pd.to_datetime(dt["trip_creation_time"])
dt["od_start_time"] = pd.to_datetime(dt["od_start_time"])
dt["od_end_time"] = pd.to_datetime(dt["od_end_time"])
```
