```python
#Testing the assumptions of the linear regression model 

#1. Multicollinearity check by VIF score 

from statsmodels.stats.outliers_influence import variance_inflation_factor

def calc_vif(dataset):

    # Calculating VIF
    vif = pd.DataFrame()
    vif["features"] = dataset.columns
    vif["VIF Value"] = [variance_inflation_factor(dataset.values, i) for i in range(dataset.shape[1])]

    return(vif)
```

```python
calc_vif(X_train)
```

```python
#Since the data is columns standardized all are VIFs are already below 5 but we shall still remove multicollinearity
```

```python
X_trainc = X_train.drop('const', axis=1)
```

```python
calc_vif(X_trainc)
```

```python
#SOP has high VIF and high p (0.918) value hence we remove it
```

```python
X_trainc
```

```python
X_trainc = X_trainc.drop('SOP', axis=1)
```

```python
calc_vif(X_trainc)
```

```python
#University Rating has high VIF and high p (0.096) value hence we remove it
```

```python
X_trainc = X_trainc.drop('University Rating', axis=1)
```

```python
calc_vif(X_trainc)
```

```python
#Rest of the features all have p value which less than 0.05 and VIFs are also less than 5 
```

```python
#Calcualting R squared for new X_train
X_trainc = sm.add_constant(X_trainc)
model = sm.OLS(y_train, X_trainc).fit()
```

```python
model.params
```

```python
model.summary()

#As we can see the R squared and adjust R squared have reduced by 0.001 by the deletion of two variables hence we know
#that these two varaibles were not useful. Now in the new dataset p values for all variables are less than 0.05
```

```python
#4. Test for homoscedasticity

#Residual plot

resp = pd.DataFrame()
resp["residual"] = residual
resp["predicted value"] = y_test_pred
sns.scatterplot(data=resp, x="predicted value", y="residual")

#As we can see there is no pattern in the residual plot hence we can say that the error is homoscedastic
```

```python
import statsmodels.stats.api as sms
from statsmodels.compat import lzip

name = ['F statistic', 'p-value']
test = sms.het_goldfeldquandt(residual, X_test)
lzip(name, test)

#Null hypothesis data is homoscedastic
#Since p value is > 0.05 we fail to reject the null hypothesis
```
